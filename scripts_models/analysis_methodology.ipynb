```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Baseline - Figure 3

```python
dfUdev = pd.read_csv('results/N10_Baseline/device_Utilization.csv')
dfUdev.drop(dfUdev.columns[-1], axis=1, inplace=True)
dfUdev.sort_values(by='S_s1', inplace=True)
dfUser = pd.read_csv('results/N10_Baseline/server_Utilization.csv')
dfUser.drop(dfUser.columns[-1], axis=1, inplace=True)
dfUser.sort_values(by='S_s1', inplace=True)
dfR0 = pd.read_csv('results/N10_Baseline/_SystemResponseTime.csv')
dfR0.drop(dfR0.columns[-1], axis=1, inplace=True)
dfR0.sort_values(by='S_s1', inplace=True)
```

```python
ciSize = (dfR0['upper'] - dfR0['lower']) / 2
fig, ax1 = plt.subplots()
l1 = ax1.plot(dfR0['S_s1'], dfR0['mean'], color='b', ls='-', label=r'$R_{sys}^{dflt}$')
ax1.fill_between(dfR0['S_s1'], dfR0['mean']-ciSize, dfR0['mean']+ciSize, color='b', alpha=.3)

ciSize = (dfUdev['upper'] - dfUdev['lower']) * 100 / 2
ax2 = ax1.twinx()
l2 = ax2.plot(dfUdev['S_s1'], dfUdev['mean']*100, color='r', ls='-.', label=r'$U_{dev}$')
ax2.fill_between(dfUdev['S_s1'], dfUdev['mean']*100-ciSize, dfUdev['mean']*100+ciSize, color='r', alpha=.3)

ciSize = (dfUser['upper'] - dfUser['lower']) * 100 / 2
l3 = ax2.plot(dfUser['S_s1'], dfUser['mean']*100, color='g', ls=':', label=r'$U_{ser}$')
ax2.fill_between(dfUser['S_s1'], dfUser['mean']*100-ciSize, dfUser['mean']*100+ciSize, color='g', alpha=.3)


ax1.set_xlabel(r'$S_{ser}$', size=14)
ax1.set_ylabel('System Response Time', size=14)
ax2.set_ylabel('Utilization [%]', size=14, rotation=270, labelpad=16)

ax1.set_yticks(np.arange(0.,0.81,0.2))
ax1.tick_params(labelsize=14)
ax2.set_yticks(range(0,101,25))
ax2.tick_params(labelsize=14)
ax1.set_ylim(0.,0.835)
ax2.set_ylim(0,106.6667)

plt.axvline(0.04, ls='--', color='k')

ls = l1+l2+l3
labs = [l.get_label() for l in ls]
ax1.legend(ls, labs, loc='upper right', bbox_to_anchor=(1,0.72), fontsize=14)

fig.tight_layout()
fig.set_size_inches(6, 4)
#plt.savefig('images/baseline_anal.pdf')
plt.show()
```
